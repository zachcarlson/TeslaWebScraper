```python
from collections import *
import csv
```

```python
def federal_state_search():
    from pprint import pprint
    fed_state_search = ''
    address_list = []
    count_yes = ['Yes', 'yes', 'y', 'YES']
    count_no = ['No', 'no', 'n', 'NO']
    federal_state_results = []
    state_abbreviations = [
    'AL','al','AK','ak','AZ','az','AR','ar','CA','ca','CO','co',
    'CT','ct','DE','de','FL','fl','GA','ga','HI','hi','ID','id',
    'IL','il','IN','in','IA','ia','KS','ks','KY','ky','LA','la',
    'ME','me','MD','md','MA','ma','MI','mi','MN','mn','MS','ma',
    'MO','mo','MT','mt','NE','ne','NV','nv','NH','nh','NJ','nj',
    'NM','nm','NY','ny','NC','nc','ND','nd','OH','oh','OK','ok',
    'OR','or','PA','pa','RI','ri','SC','sc','SD','sd','TN','tn',
    'TX','tx','UT','ut','VT','vt','VA','va','WA','wa','WV','wv',
    'WI','wi','WY','wy']
    while True:
        fed_state_search = input("What state are you looking to find chargers? (Or enter '0' to exit):\t")
        if fed_state_search == '0':
            break
        if fed_state_search in state_abbreviations:
            break
        else:
            print('Invalid input. Please enter the abbreviation of the state you are looking for.')
            fed_state_search
    
    fed_state_search = fed_state_search.upper()
    
    while True:
        if fed_state_search == '0':
            break

        with open('data/alt_fuel_stations (Oct 4 2020).csv', 'r') as file:
            stuff = csv.reader(file)
            
            for line in stuff:
                if line[5] in fed_state_search:
                    federal_state_results.append(line[4])
                    address = line[2],line[4]+', '+line[5]
                    address_list.append(address)
                    address_list.sort(key = lambda x: x[1]) 
                    
            if len(federal_state_results) > 0:
                print('Here are the street addresses of the chargers we found in ' + fed_state_search, '\n')
                pprint(address_list)
                break
                
            if len(federal_state_results) < 1:
                print("We could not find any chargers in " + fed_state_search + ". Please try again or enter '0' to exit.")




    while True:
        if fed_state_search == '0':
            break
        count_of_cities_question = input('Would you also like a count of chargers in each city?:\t')
        print('\n')
        if count_of_cities_question in count_yes:
            print('Here are those same cities in ' + fed_state_search + ', with the amount of chargers we found in that city.' '\n')
            state_charger_count = Counter(federal_state_results)
            print(state_charger_count)
            break
        if count_of_cities_question in count_no:
            break
        if count_of_cities_question not in count_yes or count_no:
            print('Invalid entry. Please try again.')
                        


                
# federal_state_search()
```

```python
def tesla_state_search():
    from pprint import pprint
    tesla_state_search = ''
    address_list = []
    count_yes = ['Yes', 'yes', 'y', 'YES']
    count_no = ['No', 'no', 'n', 'NO']
    tesla_results = []
    state_abbreviations = [
    'AL','al','AK','ak','AZ','az','AR','ar','CA','ca','CO','co',
    'CT','ct','DE','de','FL','fl','GA','ga','HI','hi','ID','id',
    'IL','il','IN','in','IA','ia','KS','ks','KY','ky','LA','la',
    'ME','me','MD','md','MA','ma','MI','mi','MN','mn','MS','ma',
    'MO','mo','MT','mt','NE','ne','NV','nv','NH','nh','NJ','nj',
    'NM','nm','NY','ny','NC','nc','ND','nd','OH','oh','OK','ok',
    'OR','or','PA','pa','RI','ri','SC','sc','SD','sd','TN','tn',
    'TX','tx','UT','ut','VT','vt','VA','va','WA','wa','WV','wv',
    'WI','wi','WY','wy']
    while True:
        tesla_state_search = input("What state are you looking to find Superchargers? (Or enter '0' to exit):\t")
        if tesla_state_search == '0':
            break
        if tesla_state_search in state_abbreviations:
            break
        else:
            print("Invalid input. Please enter the abbreviation of the state you are looking for. (Or enter '0' to exit)")
            tesla_state_search
    
    tesla_state_search = tesla_state_search.upper()
    
    while True:
        if tesla_state_search == '0':
            break

        file =  csv.reader(open('data/tesla_superchargers.csv', 'r', encoding = 'utf-8'))
        
        print('Here are the cities we found with Superchargers in ' + tesla_state_search, '\n') 
    
        for line in file:
            if tesla_state_search in line[2]:
                tesla_results.append(line[2])
                address = line[1], ', '+line[2]
                address_list.append(address)
                address_list.sort(key = lambda x: x[1])
                
        if len(tesla_results) > 0:
            pprint(address_list)
            break
        if len(tesla_results) < 1:
            print("We could not find any Superchargers in "+tesla_state_search+ ". Please try again or enter '0' to exit.")
            
            
    while True:
        if tesla_state_search == '0':
            break
        count_of_cities_question = input('Would you also like a count of chargers in this state?:\t')
        print('\n')
        
        if count_of_cities_question in count_yes:
            print('In ' + tesla_state_search + ', there are this many charging locations.' '\n')
            state_charger_count = Counter(tesla_results)
            print(state_charger_count)
            break
        if count_of_cities_question in count_no:
            break
        if count_of_cities_question not in count_yes or count_no:
            print('Invalid entry. Please try again.')
        
# tesla_state_search()
```
