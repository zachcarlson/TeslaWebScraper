```python
from bs4 import BeautifulSoup
import urllib
# We'll still need this to download webpages
from pprint import pprint
# This just downloads the html text with full markdown
html_text = urllib.request.urlopen("https://www.tesla.com/findus/list/superchargers/United%20States").read()

# We want a markdown-interpreted (structured) version of the html using BeautifulSoup:
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
street_address = soup.find_all('span', 'street-address')
locality = soup.find_all('span', 'locality')
#telephone = soup.find_all('span', 'value')
```

```python
#iterate over all spans and see what they are:
for item in street_address:
    print(item.text)
    
for item in locality:
    print(item.text)
    
# for item in telephone:
#     print(item.text)
```

```python
#read elements into lists
#remove whitespace and just pull in text from span and a elements

street_address_list = []
for item in street_address:
    if item.text == "":
        street_address_list.append("No address given")
    else:
        street_address_list.append(item.text.strip())
        
city_list = []
for item in locality:
        city_list.append(item.text.strip())
        
# telephone_list = []
# for item in telephone_number:
#         telephone_list.append(item.text.strip())
    
pprint(street_address_list)
pprint(city_list)
# pprint(telephone_list)

# print(len(street_address_list), len(city_list), len(telephone_list))

#the telephones are not aligned and if they have numbers, it's the same
#the telephones are not in our scope and not really valuable data, considering they're all the same
```

```python
#read in street-address, locality, value
#value is telephone number
title = soup.find_all('a', 'fn org url')

#checking if title can be the name of our dictionaries.  comparing length of title list and tesla_set set
#print(len(title))

# tesla_set = set()

# for item in title:
#     if item not in tesla_set:
#         tesla_set.add(item)
#     else:
#         print(item.text)
        

#print(len(tesla_set))

#:( 1198 != 1195

#three duplicates are Brunswick, GA Supercharger; Hudsonville, MI supercharger; arlington, VA supercharger
#we can attest to singular charging stations having multiple units, the duplication is probably not important and an error

#makes unique key values for duplicate names
title_list = []
for item in title:
    if item.text not in title_list:
        title_list.append(item.text.strip())
    else:
        updated_name = item.text + " #2"
        title_list.append(updated_name.strip())

# tesla_chargers = [
#     'Athens, AL Supercharger': {
#         'street-address':'21282 Athens-Limestone Blvd',
#         'locality':'Athens, AL'
#     },
#     'next_dict': { 
#          ... 
#     }
#     ...
# ]pprint(title_list)
```

```python
# tesla_chargers = [
#     'Athens, AL Supercharger': {
#         'street-address':'21282 Athens-Limestone Blvd',
#         'locality':'Athens, AL'
#     },
#     'next_dict': { 
#          ... 
#     }
#     ...
# ]


#make list of dictionaries
# dict_keys = ("street-address", "locality")
# tesla_dict = dict(zip(title_list, zip(street_address_list, city_list)))
# pprint(tesla_dict)
tesla_dict = {}
for s,c,t in zip(street_address_list, city_list, title_list):
    tesla_dict[t] = {}
    tesla_dict[t]['charging station name'] = t
    tesla_dict[t]['street address'] = s
    tesla_dict[t]['city'] = c
print(tesla_dict)
```

```python
import csv

with open("tesla_superchargers.csv", 'w') as file_handle:
    fieldnames = ['charging_station_name', 'street_address', 'city']
    writer = csv.DictWriter(file_handle, delimiter=',', lineterminator='\n', fieldnames = fieldnames)
    writer.writeheader()
    for key in tesla_dict:
        values = []
        for item, value in tesla_dict[key].items():
            values.append(value)
        writer.writerow({'charging_station_name': values[0],
                        'street_address': values[1],
                        'city': values[2]})
```
