```python
#CELL 1:
##A. Import modules and packages
##B. Scrape website 

from bs4 import BeautifulSoup
import urllib
from pprint import pprint 
import re
import csv

html_text = urllib.request.urlopen("https://www.tesla.com/findus/list/superchargers/United%20States").read()
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
#CELL 2:
##A. Using soup, get lists of street addresses, 
#title (i.e. supercharger name), and locality (i.e. city, state, and zipcode)

street_address = soup.find_all('span', 'street-address')
locality = soup.find_all('span', 'locality')
title = soup.find_all('a', 'fn org url')
```

```python
#CELL 3:
##A. Iterate over imported lists to see their contents

for item in title:
    print(item.text)

for item in street_address:
    print(item.text)
    
for item in locality:
    print(item.text)
```

```python
#CELL 4: 
##A. Tidying lists contents.

#read elements into lists
#remove whitespace and just pull in text from span and a elements

street_address_list = []
for item in street_address:
    if item.text == "":
        street_address_list.append("No address given")
    else:
        street_address_list.append(item.text.strip())
        
city_list = []
for item in locality:
        city_list.append(item.text.strip())

#makes unique key values for duplicate names
title_list = []
for item in title:
    if item.text not in title_list:
        title_list.append(item.text.strip())
    else:
        updated_name = item.text + " #2"
        title_list.append(updated_name.strip())   
        
pprint(street_address_list)
pprint(city_list)
```

```python
#CELL 5: 
##A. Combine street address, city, and title lists into a dictionary
##B. Take city information and breakdown into city, state, and zipcode

# tesla_chargers = [
#     'Athens, AL Supercharger': {
#         'street-address':'21282 Athens-Limestone Blvd',
#         'locality':'Athens, AL'
#     },
#     'next_dict': { 
#          ... 
#     }
#     ...
# ]

tesla_dict = {}
for s,c,t in zip(street_address_list, city_list, title_list):
    #initialize values and refresh
    tesla_dict[t] = {}
    city = ""
    state = ""
    zipcode = ""

    items = re.split("[,]", c) #split city and state
    city = items[0]
    state = items[1].strip()
    if len(state) > 0: #some entries have state as ""
        if state[0].isdigit(): #if first character is digit, it is the zipcode and there is no state
            zipcode = state
            state = "No state"
        elif len(state) > 2: #if state has more than two char, it contains zipcode
            zipcode = state[2:].strip()
            state = state[:2] #get first 2 characters
        else:
            zipcode = "No zipcode"
    else:
        state = "No state"
    tesla_dict[t]['charging_station_name'] = t
    tesla_dict[t]['city'] = city
    tesla_dict[t]['state'] = state
    tesla_dict[t]['zipcode'] = zipcode
    tesla_dict[t]['street_address'] = s

print(tesla_dict)
```

```python
#CELL 6:
##A. Export nested dictionary as a .csv file

with open("data/tesla_superchargers.csv", 'w') as file_handle:
    fieldnames = ['charging_station_name', 'city', 'state', 'zipcode', 'street_address']
    writer = csv.DictWriter(file_handle, delimiter=',', lineterminator='\n', fieldnames = fieldnames)
    writer.writeheader()
    for key in tesla_dict:
        values = []
        for item, value in tesla_dict[key].items():
            values.append(value)
        writer.writerow({'charging_station_name': values[0],
                         'city': values[1], 
                         'state': values[2], 
                         'zipcode': values[3], 
                         'street_address': values[4]
                        })
```
